```python
# 문제. 
# 칼로리가 200 이하이면서 가격이 5000원 이상인 메뉴를 필터링 하세요
```

```python
cond = (df["칼로리"] <= 200) & (df["가격"] >= 5000)

cond1 = df["칼로리"] <= 200
cond2 = df["가격"] >= 5000

df[cond1 & cond2]
```

```python
# NOT 연산자 

cond = df["칼로리"] < 50
df[~cond]
```

```python
# 복수 조건 필터링

cond1 = df["가격"] > 5000
cond2 = df["칼로리"] > 100

df[cond1 | cond2]
```

```python
# isin 필터링
cond = df["메뉴"].isin(["녹차"])
```

```python
df[cond]
```
