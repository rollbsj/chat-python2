```python
# max(), min(), mean(), median(), sum(), std(), var()

df
```

```python
print("최대값", df["가격"].max())
print("최소값", df["가격"].min())
print("평균",   df["가격"].mean())
print("중앙값", df["가격"].median())
print("합계",   df["가격"].sum())
print("표준편차", df["가격"].std())
print("분산",   df["가격"].var())
```

```python
# 백분위수 quantile()

print("분위수 25% 값", df["가격"].quantile(0.25))
print("분위수 75% 값", df["가격"].quantile(0.75))
```

```python
# 최빈값 mode()
df["원산지"].mode()
```

```python
# idxmax(), idxmin() 
# 데이터프레임에서 최대값을 가지고 있는 index를 구할때 
# 데이터프레임에서 최소값을 가지고 있는 index를 구할때 

print(df["가격"].idxmax())
print(df["가격"].idxmin())
```

```python
# nlargest(), nsmallest()
# 데이터프레임에서 특정 컬럼에서 가장 큰 값 
# 또는 가장 작은 값을 반환할때

df.nlargest(1, "가격")
```
