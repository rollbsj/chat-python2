```python
## 튜플은 리스트 형태와 비슷
## 리스트는 수정이 가능하고, 튜플은 수정이 불가능
```

```python
# 튜플 생성
t1 = ()
t2 = (1,) # l2 = [1]
t3 = (1, 2, 3)
t4 = 1, 2, 3

# result = 108, "많음"
t5 = ("a", "b", (1, 2))

```

```python
# 튜플 요소를 삭제

# 리스트요소 예시
list1 = [1, 2, 3]
del list1[1]
print("리스트요소삭제:", list1)

t1 = (1, 2, 3) 
del t1[1]
print("튜플요소삭제:", t1)
```

```python
# 튜플 요소를 수정 

# 리스트요소 예시
list1 = [1, 2, 3]
list1[1] = 4
print("리스트요소수정:", list1)

# 튜플요소 예시
t1 = (1, 2, 3)
t1[1] = 4
print("튜플요소수정:", t1)
```
