```python
### 비교연산자 
```

```python
# 크다 >
# 크거나같다(이상) >= 
# 작다 < 
# 작거나같다(이하) <=  
# 같다 ==  
# 다르다 != 
```

```python
# 참조건 
print(10 > 5 )
print(10 < 5)
print(10 >= 10)
print(10 <= 10)
print(10 == 10)
print(10 != 10)
print( 10 = 10)

```

```python
# 조건문의 구성 

score = 2 

if score >= 10: 
    print("score가 10이상입니다.")
elif score < 10 and score >= 5:
    print("score가 5이상 10미만입니다.")
elif score < 5 and score >= 3:
    print("score가 3이상 5미만입니다.")
else: 
    print("score가 3미만입니다.")
```
