```python
ken = {"name" : "홍길동", "gender": "male", "hobby": "running"}
print(ken["name"])
print(ken["gender"])
print(ken["hobby"])
```

```python
listbox = ken.keys()
result = list(listbox)
result 

list(ken.keys())
```

```python
list_values = ken.values()
result = list(list_values)
result 
```

```python
# key로 Value 값 얻기 - pop
ken = {'name': '홍길동', 'gender': 'male', 'hobby': 'running'}

result = ken.pop("hobby")
result 
```
