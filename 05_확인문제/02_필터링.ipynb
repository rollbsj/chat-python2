```python
import pandas as pd
# df = pd.read_csv("type1_data1.csv")
df = pd.read_csv("type1_data1.csv")
```

```python
print(df["f5"])
```

```python
# 1. 'f5' 컬럼이 0이 아닌 데이터(행)을 구하시오
cond = df["f5"] != 0
df=df[cond]
df
```

```python
# 2. 앞에서 구한 데이터에 "views" 컬럼 결측치를 "views" 컬럼의 최솟값을 채워주세요 
min2=df["views"].min()
df["views"]=df["views"].fillna(min2,inplace=True)
df
#df["views"].isnull().sum()
```

```python
#3. 그리고 "views" 컬럼의 중앙값을 계산해 정수로 구하시오
int(df["views"].median())
```
