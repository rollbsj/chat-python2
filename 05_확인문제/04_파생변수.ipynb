```python
import pandas as pd 
df = pd.read_csv("type1_data1.csv")
```

```python
df.isnull().sum()
df.shape
```

```python
# 1. 결측치가 있는 데이터(행)을 제거하시오
df=df.dropna()
df
```

```python
# 2. "views" 컬럼을 "f1" 컬럼으로 나눈 값을 새로운 컬럼으로 추가히시오 
df["추가"]=df["views"]/df["f1"]
df["추가"]
df.head(2)
```

```python
# 3. 새로운 컬럼 값 중 가장 큰 값을 가진 행의 age를 정수로 구하시오
df.sort_values("추가",ascending=False).head(1)
```

```python
result2 = df.nlargest(1,"추가")
```
