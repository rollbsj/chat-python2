```python
import pandas as pd 

# 데이터 불러오기
df = pd.read_csv("type1_data1.csv")
```

```python
df
```

```python
Q1 = df["views"].quantile(0.25)
Q3 = df["views"].quantile(0.75)
iqr = Q3 - Q1

print(Q1)
print(Q3)
print(iqr)
```

```python
low_q = Q1 -(iqr * 1.5)
upper_q= Q3-(iqr * 1.5)
cond1 = df["views"] < low_q
cond2 = df["views"] > upper_q
df=df[cond1 | cond2]
df
```

```python

# 1. "views" 컬럼의 1사분위수, 3사분위수, iqr구하시오
# (iqr = 3사분위수 - 1사분위수)
Q1 = df["views"].quantile(0.25)
Q3 = df["views"].quantile(0.75)
iqr = Q3 - Q1

print(f"1사분위수(Q1): {Q1}")
print(f"3사분위수(Q3): {Q3}")
print(f"IQR: {iqr}")

# 2. 이상치 조건에 맞는 데이터를 찾으시오 
# (이상치는 1사분위수 - (iqr *1.5)보다 작은 값고 3사분위 + (iqr * 1.5)보다 큰값)
lower_bound = Q1 - (iqr * 1.5)
upper_bound = Q3 + (iqr * 1.5)

outliers = df[(df["views"] < lower_bound) | (df["views"] > upper_bound)]
print(f"\n이상치 개수: {len(outliers)}")

# 3. 이상치 데이터의 "views" 컬럼 합을 정수로 구하시오 
outlier_views_sum = int(outliers["views"].sum())
print(f"이상치 데이터의 'views' 컬럼 합: {outlier_views_sum}")
```
