```python
import pandas as pd 
df = pd.read_csv("type1_data1.csv")

# 1. "views" 컬럼의 결측 데이터를 0으로 대체하시오.
df["views"] = df["views"].fillna(0)

# 2. "views" 컬럼에서 상위 10번째 값을 구하시오.
# 내림차순 정렬 후 10번째(인덱스 9) 값
views_top10th = df["views"].sort_values(ascending=False).iloc[9]
print("상위 10번째 값:", views_top10th)

# 3. "views" 컬럼에서 상위 10개의 값을 상위 10번째 값으로 대체하시오.
# 상위 10개 행의 인덱스를 구한 뒤, 그 값들을 모두 views_top10th로 변경
idx_top10 = df["views"].sort_values(ascending=False).index[:10]
df.loc[idx_top10, "views"] = views_top10th

# 4. "views" 컬럼의 전체 합을 정수로 구하시오.
views_sum = int(df["views"].sum())
print("views 합(정수):", views_sum)
```
