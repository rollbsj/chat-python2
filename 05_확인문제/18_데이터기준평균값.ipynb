```python
import pandas as pd 

# 데이터 불러오기
df = pd.read_csv("type1_data2.csv", index_col=0)

# 1. index "2001" 데이터(행)에서 평균보다 큰 값의 개수를 구하시오 
# 2. index "2003" 데이터(행)에서 평균보다 작은 값의 개수를 구하시오 
# 두 개수를 더하시오. 
```

```python
df.head()
```

```python
mean_up=df.loc[2001].mean() < df.loc[2001]
mean_up.sum()
```

```python
mean_low=df.loc[2003].mean() > df.loc[2003]
mean_low.sum()
```
