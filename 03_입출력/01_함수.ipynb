```python
# 일반적인 함수

# def 함수이름(매개변수):
#     수행할 문장
#     return 반환값

def add(a, b):
    result = a + b
    return result 
```

```python
c = add(10, 20)
print(c)


print("hi")
```

```python
# 입력값이 없는 함수 

def say():
    return "Hello" 

a = say()
print(a)
```

```python
# 반환값이 없는 함수 
def add(a, b): 
    print(f"{a} + {b} = {a + b}")


result = add(10, 20)
print(result)
```

```python
# 입력값도 없고 반환값도 없는 함수 
def say():
    print("Hello")

result = say()
print(result)
```

```python
# 매개변수를 지정하여 호출하기 
def sub(a, b): 
    return a - b


result = sub(a=10, b=2)
print(result)
```

```python
result = sub(b=5, a=3)
print(result)
```

```python
# 입력값이 몇 개가 될지 모를때 어떻게 함수를 만들지?
def add_many(*args): 
    result = 0
    for i in args: 
        result = result + i
    return result 

hap = add_many(1, 2, 3, 4, 5) 
print(hap)
```

```python
def add_mul(choice, *args): 
    if choice == "add": 
        result = 0 
        for i in args: 
            result = result + i 
    elif choice == "mul": 
        result = 1 
        for i in args: 
            result = result * i 

    return result 


c = add_mul("add", 10, 30, 40)
print(c)
d = add_mul(choice="mul", 5, 3, 2)
```

```python
c = add_mul("mul", 5, 3, 2)
print(c)
```

```python
# 함수의 값은 언제나 하나이다 

def add_and_mul(a, b): 
    return a+b, a*b

result = add_and_mul(10, 20)
print(result) 
```
