```python
# 다양한 조건을 판단하는 elif 

pocket = ["paper", "cellphone"]
card = False

if "money" in pocket: 
    print("택시를 타고 가라")
else: 
    if card: 
        print("택시를 타고 가라2")
    else: 
        print("걸어가라")

```

```python
pocket = ["paper", "cellphone"]
card = False

if "money" in pocket: 
    print("택시를 타고 가라")
elif card: 
    print("택시를 타고 가라2")
else: 
    print("걸어가라")
```

```python
# 홍길도의 시험 점수 
score = 80 

# 컴퓨터시험을 쳤는데 
# 90점이상이면 "A", 
# 80점이상이면 "B", 
# 70점이상이면 "C", 
# 나머지는 "F" 

if score >= 90:
    print("A") 
elif score >= 80: 
    print("B")
elif score >= 70: 
    print("C")
else: 
    print("F")

```

```python
# 변수 = 참일 때 값 if 조건 else 거짓일 때 값 

age = 19
status = "성인" if age >= 18 else "미성년"
print(status)

if age >= 18: 
    print("성인")
else: 
    print("미성년")
```
